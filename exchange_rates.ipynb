# Exchange rate between the euro and the American dollar

## Data cleaning

We will follow these steps:

* We will rename [US dollar ] and Period\Unit columns in order to make them easier to work with.
* We will change the Time column to a datetime data type.
* The US dollar column have non-Null values, but it is classified as an object instead of a numerical column. We have to modify the column and convert it into a float64.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar','Period\\Unit:': 'Time'},inplace=True)
# We rename the two mentioned columns.
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
# We modify the Time column to datetime type:
# In pandas, you can use the pd.to_datetime() function to convert a column to a datetime object 
# without explicitly importing the datetime module. 
exchange_rates.sort_values('Time', inplace=True)
# We sort the values by Time to check easily the evolution of the rates.
exchange_rates.reset_index(drop=True, inplace=True)
# The line exchange_rates.reset_index(drop=True, inplace=True) is used to reset the index of the DataFrame exchange_rates after sorting it.
# By default, when you sort a DataFrame, the index values are preserved and correspond to the original order of the rows. 
# However, sometimes it's desirable to reset the index to have a new sequential index that reflects the new sorted order.
exchange_rates[['US_dollar','Time']].head(20)
# If we want to save our changes in the DataFrame we need to specify inplace= True. 
# When inplace=True is not specified (or inplace=False explicitly), 
# the pandas functions return a modified copy of the DataFrame instead of modifying the original object. 
# This means that if you don't assign the returned value to a variable or the original object, the changes won't be saved.
```

```python
euro_to_dollar = exchange_rates[['US_dollar','Time']]
# We now isolate the columns of our interest: US_dollar and Time.
```

```python
euro_to_dollar['US_dollar'].value_counts().head(20)
# We identify the cause for the object type of the US_dollar column:
# There are 62 rows with only a '-' symbol instead of a number.
#euro_to_dollar['US_dollar'].value_counts().tail(50)
```

```python
euro_to_dollar.drop(euro_to_dollar[euro_to_dollar['US_dollar'] == '-'].index, inplace = True)# we will drop the rows with '-' values.

# Adding .index after euro_to_dollar[euro_to_dollar['US_dollar'] == '-'] ensures that we are passing 
# the index labels of the rows to the drop() method, rather than the entire rows themselves. 
# This will avoid the KeyError we encountered and drop the rows correctly.

# This is another way to do it: euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'].value_counts().head()
```

```python
euro_to_dollar.info()
```

Now we kept 5637 non-null rows but we need to convert the US_dollar column into a numerical one:

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

## Visualizing the evolution of the Euro-Dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Enables Jupyter to display graphs

plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title('Evolution of the euro-dollar exchange rate')
plt.xlabel('Year')
plt.ylabel('Euro-Dollar rate')
plt.show()
```

As we observe in the plot the rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

In order to show only the long-term trends, we can use the rolling mean (also known as the moving average): we will use a rolling window of 30 days.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

# The pandas Series.rolling().mean() method calculates the rolling mean for each day using a specified window size. For instance, if the window size is set to 2, which means the average is calculated by taking the mean of the current day and the previous day.
# The rolling window can take various values — some common values include seven (one week), 30 (one month), 50 or 100 days, or 365 days (one year).
#print(euro_to_dollar['rolling_mean'])
plt.plot(euro_to_dollar['Time'],euro_to_dollar['rolling_mean'])
plt.title('Evolution of the euro-dollar exchange rate')
plt.xlabel('Year')
plt.ylabel('Euro-Dollar rate')

plt.show()
```

We are going to compare the following crisis that might have affected the rate between US dollar and Euro: 

1. Global Financial Crisis (2008-2009): The financial crisis, triggered by the mortgage crisis in the United States, had repercussions worldwide and affected economic stability and exchange rates between various currencies, including the dollar and the euro.

2. Eurozone Sovereign Debt Crisis (2009-2012): Several Eurozone countries faced debt problems, which generated uncertainty in financial markets and affected the exchange rate between the euro and the dollar.

3. COVID-19 Pandemic (from 2020): The global health crisis and the confinement measures adopted worldwide due to COVID-19 had a significant impact on economies and exchange rates, including the relationship between the dollar and the euro.
